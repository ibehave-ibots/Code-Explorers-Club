Now we have many related functions defined 
To have control over attributes and methods we combine them in a class

```python
class Calculator:
    def __init__(self,a,b):
        print(f"Setting self.a = {a}") # attribute
        print(f"Setting self.b = {b}") # attributes
        self.a = a
        self.b = b
        
    def add(self):                    # method
        print(f"Adding self.a:{self.a} self.b:{self.b}") 
        return self.a+self.b
    
    def sub(self):                    # method
        print(f"Subtracting self.a:{self.a} self.b:{self.b}")
        return self.a-self.b
    
    def mult(self):                    # method
        print(f"Multiplying  self.a:{self.a} self.b:{self.b}")
        return self.a*self.b
    
    def div(self):                      # method
        print(f"dividing self.a:{self.a} self.b:{self.b}")
        return self.a/self.b
```

```python
obj1 = Calculator(7,3)
```

```python
### let's find out to which class our created character belongs to
print(type(obj1))
```

```python
# Perform operations on calc1
print("Addition:", obj1.add())
print("Subtraction:", obj1.sub())
```

```python
calc2 = Calculator(20, 4) #instantiation
```

```python
# Perform operations on calc2
print("Addition:", calc2.add())
print("Subtraction:", calc2.sub())
```

We create two Calculator objects, obj1 and obj2, each with its own a and b attributes.
We perform various arithmetic operations using the methods defined in the Calculator class.
Each instance of Calculator maintains its own state (values of a and b), thanks to the self parameter in the methods and the __init__ constructor.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class DataAnalysis:
    def __init__(self, df):
        self.df = df

    def check_missing_values(self):
        print("\nMissing values in the dataset:")
        print(self.df.isnull().sum())

    def plot_class_distribution(self):
        plt.figure(figsize=(8, 6))
        self.df['target'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
        plt.title('Distribution of Diagnosis Classes')
        plt.xlabel('Diagnosis')
        plt.ylabel('Count')
        plt.xticks(ticks=[0, 1], labels=['Benign', 'Malignant'], rotation=0)
        plt.show()

    def plot_correlation_heatmap(self):
        plt.figure(figsize=(12, 10))
        sns.heatmap(self.df.corr(), cmap='coolwarm', fmt='.2f', linewidths=0.5)
        plt.title('Correlation Heatmap')
        plt.show()
```

```python
from sklearn.datasets import load_breast_cancer, load_iris

# Load breast cancer dataset
breast_cancer = load_breast_cancer()
breast_cancer_df = pd.DataFrame(data=breast_cancer.data, columns=breast_cancer.feature_names)
breast_cancer_df['target'] = breast_cancer.target

# Load iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python
iris_df['target'].value_counts()
```

```python
# Instantiate DataAnalysis class for breast cancer dataset
breast_cancer_analysis = DataAnalysis(breast_cancer_df)
```

```python
# Perform tasks on breast cancer dataset
breast_cancer_analysis.check_missing_values()
breast_cancer_analysis.plot_class_distribution()
breast_cancer_analysis.plot_correlation_heatmap()
```

```python
# Instantiate DataAnalysis class for iris dataset
iris_analysis = DataAnalysis(iris_df)
```

```python
# Perform tasks on iris dataset
iris_analysis.check_missing_values()
iris_analysis.plot_class_distribution()
iris_analysis.plot_correlation_heatmap()
```
